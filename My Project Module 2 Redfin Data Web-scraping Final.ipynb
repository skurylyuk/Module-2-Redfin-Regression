```python
def grab_info(response):
    soup = bs(response.text, 'html.parser')
    l = soup.findAll(class_='header')
    properties_list=["Price/Sq.Ft.", "Lot Size", "Year Built"]
    dic = {}

    for tag in l:
        if tag.text in properties_list:
            dic[tag.text]=tag.findNextSibling().text.replace(",", "").replace("$", "").strip()

    return dic
```

```python
def tax_info_fun(response):
    soup = bs(response.text, 'html.parser')
    tax_info = soup.findAll(class_ ='listHeader')
#list of title we want to scrape
    info = ['Tax Information']

#dict to hold the key/value pairs of the scraped info
    tax_dict = {}

#loop over all the listHeader tags
    for tag in tax_info:

    #check if the text is one of the things we care about
        if tag.text in info:

        #If it is, we loop over all the bullets in the list
            for list_item in tag.findNextSibling():
                #grab the text of the bullet item
                list_item_text = list_item.text
                #Split the string text on ":" and grab the parts we'll use as key and value
                key = list_item_text.split(":")[0].replace(",", "").replace("$", "").strip()
                value = list_item_text.split(":")[1].replace(",", "").replace("$", "").strip()

                tax_dict[key] = value
    return tax_dict
```
